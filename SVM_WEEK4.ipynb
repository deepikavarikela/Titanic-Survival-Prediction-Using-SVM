```python
#DATA CONVERSION
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
titanic.dtypes
```

```python
#convert sex attributes values to numeric
titanic['sex'] = le.fit_transform(titanic['sex'])
```

```python
titanic['sex']
```

```python
#convert sex attributes values to numeric
titanic['embarked'] = le.fit_transform(titanic['embarked'])
titanic['class'] = le.fit_transform(titanic['class'])
titanic['deck'] = le.fit_transform(titanic['deck'])
titanic['embark_town'] = le.fit_transform(titanic['embark_town'])
titanic['alive'] = le.fit_transform(titanic['alive'])
titanic['alone'] = le.fit_transform(titanic['alone'])
titanic['who'] = le.fit_transform(titanic['who'])
```

```python
titanic
```

```python
#DATA SEPRATION
Y = titanic['survived']
```

```python
Y
```

```python
X = titanic.drop(['survived'],axis = 1)
```

```python
X
```

```python
#DATA SPLITTING
from sklearn.model_selection import train_test_split
```

```python
#70 % training and 30% testing
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

```python
#model selection - LINEAR SVM
from sklearn.svm import SVC
```

```python
lin_svm = SVC()
```

```python
#model training
lin_svm.fit(X_train,Y_train)
```

```python
#model prediction
lin_svm_pred = lin_svm.predict(X_test)
```

```python
lin_svm_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("accuracy score of linear svm ",accuracy_score(Y_test,lin_svm_pred)*100)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print("confusion matrix of linear svm \n",confusion_matrix(Y_test,lin_svm_pred))
```

```python
#2nd model polynomial svm
poly_svm=SVC(kernel = 'poly')
```

```python
poly_svm.fit(X_train,Y_train)
```

```python
poly_pred = poly_svm.predict(X_test)
```

```python
poly_pred
```

```python
print("accuracy of polynomial svm",accuracy_score(Y_test,poly_pred)*100)
```

```python
print("confusion matrix \n",confusion_matrix(Y_test,poly_pred))
```
